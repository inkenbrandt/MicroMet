```python
import requests
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

station = '(138,139,140,141)'
startdate = '2024-04-01'

headdict = {'Accept-Profile': 'gwportal','Content-Type': 'application/json'}

params = {'locationid':f'in.{station}',
         'readingdate':f'gte.{startdate}'
         }

resp = requests.get("https://ugs-koop-umfdxaxiyq-wm.a.run.app/reading",headers=headdict,params=params)

df = pd.DataFrame(resp.json())
df['readingdate'] = pd.to_datetime(df['readingdate'],errors='coerce')
df.drop_duplicates(subset=['locationid','readingdate'],inplace=True,)
df.set_index(['locationid','readingdate'],inplace=True)
df = df['waterelevation'].unstack(level=0).resample('1h').mean()

# Create a figure with secondary y-axis enabled
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Trace for the first dataset (primary y-axis)
fig.add_trace(
    go.Scatter(
        y=df[138],
        name="Dataset A",
        mode="lines"
    ),
    secondary_y=False
)

# Trace for the second dataset (secondary y-axis)
fig.add_trace(
    go.Scatter(
        y=df[139],
        name="Dataset B",
        mode="lines"
    ),
    secondary_y=True
)

# Update layout for titles, etc.
fig.update_layout(
    title="Two Time Series with Two Y-Axes",
    xaxis_title="Date",
    yaxis_title="Primary Y-Axis (Dataset A)",
)
fig.update_yaxes(title_text="Secondary Y-Axis (Dataset B)", secondary_y=True)

fig.show()
```
